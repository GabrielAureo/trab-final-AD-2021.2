```python
import estruturas_arv
import finitas

# Parâmetros para a questão 5.2
# Os parâmetros iniciais dão os melhores resultados mas são extremamente demorados.
# Para testes mais rápidos é recomendado usar no máximo 200 runs e 2000 de ḿax_time 
# e max_events, mas nesse caso não é garantido que os resultados analíticos vão 
# estar dentro do intervalo dos intervalos de confiança.
runs=300
max_time=3000
max_events=3000
minimoParaInfinitude=300

# Função para rodar todas os items da questão 5.2
def executar(lamda,mu,kind):
    arvores = estruturas_arv.gerarArvores(lamda,mu,kind,runs,max_time,max_events,minimoParaInfinitude)
    
    estruturas_arv.q1numerica(lamda,mu,kind,runs,max_time,max_events)
    estruturas_arv.q1analitica(lamda,mu,kind)
    print()
    estruturas_arv.q2numerica(arvores)
    estruturas_arv.q2analitica(lamda,mu,kind)
    print()
    estruturas_arv.q3numerica(arvores)
    print()
    estruturas_arv.q4numerica(arvores)
    estruturas_arv.q4analitica(lamda,mu,kind)
    print()
    estruturas_arv.q5numerica(arvores)
    estruturas_arv.q5analitica(lamda,mu,kind)
    print()
    estruturas_arv.q6numerica(lamda,mu,kind,runs,max_time,max_events,minimoParaInfinitude)
    estruturas_arv.q6analitica(lamda,mu,kind)
    print()
    estruturas_arv.q7numerica(arvores)
    estruturas_arv.q7analitica(lamda,mu,kind)
```
